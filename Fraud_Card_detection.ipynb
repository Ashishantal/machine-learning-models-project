```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import  accuracy_score
```

```python
df = pd.read_csv("/content/creditcard.csv")
```

```python
new = df.dropna()
```

```python
new.isnull().sum()
```

```python
new.describe()
```

```python
legit=new[new.Class ==0]
fake =new[new.Class==1]
```

```python
legit.shape
```

```python
fake.shape
```

```python
#compare legit and fake
new.groupby('Class').mean()
```

```python
#sample are to large son we take some part from data as sample  based with fake as 2 ratio 1
ls = legit.sample(n=100)
```

```python
newd = pd.concat([ls,fake],axis=0)
```

```python
newd.head()
```

```python
newd.info()
```

```python
newd['Class'].value_counts()
```

```python
newd.groupby('Class').mean()
```

```python
#traing and testing
X = newd.drop(columns ='Class',axis = 1)
Y= newd['Class']
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2 ,stratify=Y,random_state=2)
```

```python
X.shape
```

```python
X_train.shape
```

```python
#model training using logistic regression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
Ytestdata =model.predict(X_test)
acc = accuracy_score(Ytestdata,Y_test)
```
